### Soru 1: Bir öğrencinin ara sınav notunun tahminlenmesi (40 puan)

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, KFold, cross_val_score
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import mean_squared_error, mean_squared_error, r2_score,accuracy_score, classification_report, silhouette_score
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsClassifier
from sklearn.mixture import GaussianMixture
from sklearn.cluster import AgglomerativeClustering, KMeans
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)  
# Donem Sonu.xls dosyasını yükleme
data = pd.read_excel("Donem Sonu.xls")

# Kategorik Sütunların listesi
kategorik_sutunlar = ["Kayıt Şekli", "Proje1 tek/grup", "Proje2 tek/grup" ]

# One-Hot encoding için Dummy değişkenler oluşturma
df_with_dummies = pd.get_dummies(data, columns=kategorik_sutunlar)

df_with_dummies['Kayıt Şekli_ilk'] = df_with_dummies['Kayıt Şekli_ilk'].apply(lambda x: 1 if x == True else 0)
df_with_dummies['Kayıt Şekli_tekrar'] = df_with_dummies['Kayıt Şekli_tekrar'].apply(lambda x: 1 if x == True   else 0)
df_with_dummies['Proje1 tek/grup_grup'] = df_with_dummies['Proje1 tek/grup_grup'].apply(lambda x: 1 if x == True else 0)
df_with_dummies['Proje1 tek/grup_tek'] = df_with_dummies['Proje1 tek/grup_tek'].apply(lambda x: 1 if x == True   else 0)
df_with_dummies['Proje2 tek/grup_grup'] = df_with_dummies['Proje2 tek/grup_grup'].apply(lambda x: 1 if x == True else 0)
df_with_dummies['Proje2 tek/grup_tek'] = df_with_dummies['Proje2 tek/grup_tek'].apply(lambda x: 1 if x == True   else 0)

print(df_with_dummies)
# Kategorik olmayan sütunları seçme
kategorik_olmayan_sutunlar = ["Lab1 Quiz Notu", "Lab1 Prg Notu", "Lab2 Quiz Notu", "Lab2 Prg Notu",
                             "Lab3 Quiz Notu", "Lab3 Prg Notu", "Lab4 Quiz Notu", "Lab4 Prg Notu",
                             "Lab5 Quiz Notu", "Lab5 Prg Notu", "Lab6 Quiz Notu", "Lab6 Prg Notu",
                             "Lab7 Quiz Notu", "Proje1 Notu", "Proje1 Geç Gün Say", "Arasınav Notu",
                             "Lab8 Quiz Notu", "Lab8 Prg Notu","Lab9 Quiz Notu", "Lab9 Prg Notu","Lab10 Quiz Notu",
                              "Lab10 Prg Notu", "Proje2 Notu","Proje2 Geç Gün Say",]

# Min-Max normalizasyonu için scaler oluşturma
scaler = MinMaxScaler()

# Kategorik olmayan sütunları normalize edinme
normalized_data = scaler.fit_transform(df_with_dummies[kategorik_olmayan_sutunlar])

# Normalize edilmiş verileri DataFrame olarak güncelleme
df_with_dummies[kategorik_olmayan_sutunlar] = normalized_data

# Tahminlemek istenilen ara sınav notu için kullanılacak öznitelikleri belirleme
istenenler = ["Kayıt Şekli_ilk","Kayıt Şekli_tekrar", "Lab1 Quiz Notu", "Lab1 Prg Notu", "Lab2 Quiz Notu", "Lab2 Prg Notu",
                     "Lab3 Quiz Notu", "Lab3 Prg Notu", "Lab4 Quiz Notu", "Lab4 Prg Notu",
                     "Lab5 Quiz Notu", "Lab5 Prg Notu", "Lab6 Quiz Notu", "Lab6 Prg Notu",
                     "Lab7 Quiz Notu", "Proje1 Notu", "Proje1 Geç Gün Say", "Proje1 tek/grup_grup","Proje1 tek/grup_tek", "Arasınav Notu"]

# Veri setini ilgili öznitelikleri içerecek şekilde güncelleme
before_midterms = df_with_dummies[istenenler]

# Güncellenmiş DataFrame'i yazdırma
before_midterms
```

```python
#Hangi sütunların sonuca daha fazla etki ettiğini gösterme
correlation_matrix = before_midterms.corr()
correlations_with_y = correlation_matrix["Arasınav Notu"].sort_values(ascending=False)
print("'Arasınav Notu' sütunu için koralasyon değerleri: ")
print(correlations_with_y[1:])


# Bağımsız değişkenler ve hedef değişkeni ayırma
X = before_midterms.drop("Arasınav Notu", axis=1)
y = before_midterms["Arasınav Notu"]

# Veri setini eğitim ve test kümelerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Lineer Regresyon modelini oluşturun ve eğitme
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Eğitim verisi üzerinde tahmin yapma
y_train_pred = linear_model.predict(X_train)

# Test verisi üzerinde tahmin yapma
y_test_pred = linear_model.predict(X_test)

# Eğitim ve test hata değerlerini hesaplama
train_error = mean_squared_error(y_train, y_train_pred)
test_error = mean_squared_error(y_test, y_test_pred)
print("Lineer Regresyon:")
print("Eğitim Hata Değeri:", train_error)
print("Test Hata Değeri:", test_error)
linear_scores = cross_val_score(linear_model, X, y, cv=10, scoring='neg_mean_squared_error')
linear_rmse_scores = -linear_scores
linear_r2_scores = cross_val_score(linear_model, X, y, cv=10, scoring='r2')
print("Lineer Regresyon:")
print("RMSE Scores:", linear_rmse_scores)
print("Mean RMSE:", linear_rmse_scores.mean())
print("R2 Scores:", linear_r2_scores)
print("Mean R2:", linear_r2_scores.mean())
```

```python
# SVR modelini oluşturma ve hiperparametre araması yapma 

svr_model = SVR()

svr_params_a = {
    'C': [0.1, 1, 10],
    'epsilon': [0.01, 0.1, 1]
}

svr_params = {
    'C': [0.1, 1, 10, 100],  # C parametresi (hatayı tolere etme)
    'epsilon': [0.01, 0.1, 1, 10]  # Epsilon parametresi (tolerans)
}
svr_params_b = {
    'C': [0.1, 1, 10, 100],
    'epsilon': [0.01, 0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly']
}
svr_grid = GridSearchCV(svr_model, svr_params, cv=5)
svr_grid.fit(X_train, y_train)
svr_best_model = svr_grid.best_estimator_
```

```python
# K fold işlemi
svr_scores = cross_val_score(svr_model, X, y, cv=5, scoring='neg_mean_squared_error')
svr_rmse_scores = -svr_scores
svr_r2_scores = cross_val_score(svr_model, X, y, cv=5, scoring='r2')
print("SVR:")
print("RMSE Skorları:", svr_rmse_scores)
print("Mean RMSE:", svr_rmse_scores.mean())
print("R2 Skorları:", svr_r2_scores)
print("Mean R2:", svr_r2_scores.mean())
```

```python
# SVR için en iyi model Başarı hesaplama
svr_best_model = svr_grid.best_estimator_
svr_train_error = mean_squared_error(y_train, svr_best_model.predict(X_train))
svr_test_error = mean_squared_error(y_test, svr_best_model.predict(X_test))
print("SVR:")
print("Eğitim Hata Değeri:", svr_train_error)
print("Test Hata Değeri:", svr_test_error)
# SVR modelinin R2 skorunu hesaplayın
svr_r2_train = r2_score(y_train, svr_best_model.predict(X_train))
svr_r2_test = r2_score(y_test, svr_best_model.predict(X_test))
print("SVR:")
print("Eğitim R2 Skoru:", svr_r2_train)
print("Test R2 Skoru:", svr_r2_test)
```

```python
# Gradient Boosting modelini oluşturma ve eğitme
gb_model = GradientBoostingRegressor(random_state=42)
gb_params = {
'n_estimators': [50, 100, 200],
'learning_rate': [0.1, 0.05, 0.01],
'max_depth': [3, 5, 10],
'min_samples_split': [2, 5, 10],
'min_samples_leaf': [1, 2, 4],
'max_features': [None, 'sqrt', 'log2']
}

gb_grid = GridSearchCV(gb_model, gb_params, cv=5, scoring='r2')
gb_grid.fit(X_train, y_train)

# En iyi parametreleri kullanarak modeli değerlendirme
gb_best_model = gb_grid.best_estimator_
```

```python
# K fold işlemi
gb_scores = cross_val_score(gb_model, X, y, cv=5, scoring='neg_mean_squared_error')
gb_rmse_scores = -gb_scores
gb_r2_scores = cross_val_score(gb_model, X, y, cv=5, scoring='r2')

print("Gradient Boosting:")
print("RMSE Skorları:", gb_rmse_scores)
print("Mean RMSE:", gb_rmse_scores.mean())
print("R2 Skorları:", gb_r2_scores)
print("Mean R2:", gb_r2_scores.mean())
```

```python
# Gradiant Boosting için en iyi model Başarı hesaplama
gb_train_error = mean_squared_error(y_train, gb_best_model.predict(X_train))
gb_test_error = mean_squared_error(y_test, gb_best_model.predict(X_test))

print("Gradient Boosting:")
print("Eğitim Hata Değeri:", gb_train_error)
print("Test Hata Değeri:", gb_test_error)

# Gradient Boosting modelinin R2 skorunu hesaplama
gb_r2_train = r2_score(y_train, gb_best_model.predict(X_train))
gb_r2_test = r2_score(y_test, gb_best_model.predict(X_test))

print("Gradient Boosting:")
print("Eğitim R2 Skoru:", gb_r2_train)
print("Test R2 Skoru:", gb_r2_test)
```

```python
# RandomForestRegressor modelini oluşturma ve hiperparametre araması yapma
rf_model = RandomForestRegressor(random_state=42)

rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10], #  # Ağaç sayısı
    'min_samples_split': [2, 5, 10],  # Maksimum ağaç derinliği
    'min_samples_leaf': [1, 2, 4],  # Minimum örnekleme sayısı
    'max_features': [None, 'sqrt','log2']
}
rf_grid = GridSearchCV(rf_model, rf_params, cv=5, scoring="r2")
rf_grid.fit(X_train, y_train)

# En iyi parametreleri kullanarak modelleri değerlendirme
rf_best_model = rf_grid.best_estimator_
```

```python
# K fold işlemi
rf_scores = cross_val_score(rf_best_model, X, y, cv=5, scoring='neg_mean_squared_error')
rf_rmse_scores = -rf_scores
rf_r2_scores = cross_val_score(rf_best_model, X, y, cv=5, scoring='r2')

print("Random Forest:")
print("RMSE Skorları:", rf_rmse_scores)
print("Mean RMSE:", rf_rmse_scores.mean())
print("R2 Skorları:", rf_r2_scores)
print("Mean R2:", rf_r2_scores.mean())
```

```python
# Random Forest için en iyi model Başarı hesaplama
rf_train_error = mean_squared_error(y_train, rf_best_model.predict(X_train))
rf_test_error = mean_squared_error(y_test, rf_best_model.predict(X_test))

print("Random Forest:")
print("Eğitim Hata Değeri:", rf_train_error)
print("Test Hata Değeri:", rf_test_error)

# RandomForestRegressor modelinin R2 skorunu hesaplama
rf_r2_train = r2_score(y_train, rf_best_model.predict(X_train))
rf_r2_test = r2_score(y_test, rf_best_model.predict(X_test))

print("Eğitim R2 Skoru:", rf_r2_train)
print("Test R2 Skoru:", rf_r2_test)
```

### Soru 2: Bir öğrencinin final sınavında baraja takılıp takılmayacağının tahminlenmesi:  (40 puan)

```python
#Öğrenci No sütunun çıkarılması
before_finals = df_with_dummies.drop("Öğrenci No", axis=1)

#Hangi sütunların sonuca daha fazla etki ettiğini gösterme
correlation_matrix = before_finals.corr()
correlations_with_y = correlation_matrix["Final Sınavı Notu"].sort_values(ascending=False)
print("'Final Sınavı Notu' sütunu için koralasyon değerleri: ")
print(correlations_with_y[1:])

# Bağımsız değişkenler ve hedef değişkeni ayırma
X_2 = before_finals.drop("Final Sınavı Notu", axis=1)
y_2 = before_finals["Final Sınavı Notu"]
y_2 = y_2 = y_2.apply(lambda x: 0 if x < 45 else 1)

# Veri setini eğitim ve test kümelerine ayırma
X_2train, X_2test, y_2train, y_2test = train_test_split(X_2, y_2, test_size=0.3, random_state=42)

before_finals
```

```python
# RandomForestClassifier modelini oluşturun ve hiperparametre araması yapma
rf_fin_model = RandomForestClassifier(random_state=42)

rf_fin_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10], #  # Ağaç sayısı
    'min_samples_split': [2, 5, 10],  # Maksimum ağaç derinliği
    'min_samples_leaf': [1, 2, 4],  # Minimum örnekleme sayısı
    'max_features': [None, 'sqrt','log2']
}
rf_fin_grid = GridSearchCV(rf_fin_model, rf_fin_params, cv=5, scoring="r2")
rf_fin_grid.fit(X_2train, y_2train)

# En iyi parametreleri kullanarak modelleri değerlendirme
rf_fin_best_model = rf_fin_grid.best_estimator_
rf_fin_train_error = mean_squared_error(y_2train, rf_fin_best_model.predict(X_2train))
rf_fin_test_error = mean_squared_error(y_2test, rf_fin_best_model.predict(X_2test))

print("Random Forest:")
print("Eğitim Hata Değeri:", rf_fin_train_error)
print("Test Hata Değeri:", rf_fin_test_error)


y_2pred = rf_fin_best_model.predict(X_2test)

basarı = accuracy_score(y_2test, y_2pred)
report = classification_report(y_2test, y_2pred)

print("Başarıo:", basarı)
print("Sınıflandırma Raporu:\n", report)
```

```python
#Lojistik Regresyon denemesi
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_2train, y_2train)


# Test verisi üzerinde tahmin yapma
y_2_pred = log_model.predict(X_2test)

# Test verisi üzerinde tahmin yapma
accuracy = accuracy_score(y_2test, y_2_pred)
report = classification_report(y_2test, y_2_pred)
print("Logistic Regression: ")
print("Başarı:", accuracy)
print("Sınıflandırma Raporu:\n", report)
```

```python
# K-Nearest Neighbors (KNN) modelini oluşturun ve hiperparametre araması yapma
knn_fin_model = KNeighborsClassifier()

knn_fin_params = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree'],
    'p': [1, 2]
}
knn_fin_grid = GridSearchCV(knn_fin_model, knn_fin_params, cv=5, scoring="accuracy")
knn_fin_grid.fit(X_2train, y_2train)

# En iyi parametreleri kullanarak modelleri değerlendirme
knn_fin_best_model = knn_fin_grid.best_estimator_
knn_fin_train_error = mean_squared_error(y_2train, knn_fin_best_model.predict(X_2train))
knn_fin_test_error = mean_squared_error(y_2test, knn_fin_best_model.predict(X_2test))

print("K-Nearest Neighbors (KNN):")
print("Eğitim Hata Değeri:", knn_fin_train_error)
print("Test Hata Değeri:", knn_fin_test_error)
# En iyi Model ile tahminleme yapılır.
_y_2pred = knn_fin_best_model.predict(X_2test)

basarı = accuracy_score(y_2test, _y_2pred)
report = classification_report(y_2test, y_2pred)
print("Başarı:", basarı) 
print("Sınıflandırma Raporu:\n", report)
```

```python
# Kümeleme için kullanılacak sütunlar (Toplam not ortalamasını en fazla etkilediği düşünülen değerler alındı)
cluster_columns = ["Final Sınavı Notu", "Arasınav Notu", "Proje2 Notu", "Proje1 Notu"]
with_cluster = before_finals.copy()
# Final Notunu Normalize et
min_value = data["Final Sınavı Notu"].min()
max_value = data["Final Sınavı Notu"].max()
with_cluster["Final Sınavı Notu"] = (with_cluster["Final Sınavı Notu"] - min_value) / (max_value - min_value)
with_cluster2 = with_cluster.copy()
with_cluster3 = with_cluster.copy()
```

```python
# Inertia değerlerini saklamak için bir liste oluşturun
inertias = []

# Farklı küme sayıları için K-means algoritmasını çalıştırın ve inertiaları kaydedin
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init="auto")
    kmeans.fit(with_cluster[cluster_columns])
    inertias.append(kmeans.inertia_)

# Dirsek Kuralını kullanarak optimum küme sayısını belirleyin
plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Küme Sayısı')
plt.ylabel('Inertia')
plt.title('Dirsek Kuralı - Inertia Değerleri')
plt.show() 
# Dirsek kuralına göre 4 değeri gibi görünüyor
```

```python
silhouette_scores = []
silhouette_scores2 = []
# Farklı küme sayıları için K-means ve agglomerative algoritmasını çalıştırılır ve Silhouette grafikerlerini yazdırır
for k in range(3, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init="auto")
    cluster_labels = kmeans.fit_predict(with_cluster[cluster_columns])
    silhouette_scores.append(silhouette_score(with_cluster[cluster_columns], cluster_labels))
    agglomerative = AgglomerativeClustering(n_clusters=k)
    clusters = agglomerative.fit_predict(with_cluster2[cluster_columns])
    silhouette_scores2.append(silhouette_score(with_cluster2[cluster_columns], clusters))
```

```python
def Kmeans(column1, column2, a, b): # Kümeleme sonuçlarını görselleştirme
    plt.figure(figsize=(8, 6))
    for cluster_label in range(4):
        cluster_data = with_cluster[with_cluster["Küme"] == cluster_label]
        plt.scatter(cluster_data[column1], cluster_data[column2], label=f'Küme {cluster_label}')
    # Küme merkezlerini görselleştirme
    cluster_centers = kmeans.cluster_centers_
    
    plt.scatter(cluster_centers[:, a], cluster_centers[:, b], color='black', marker='X', s=100, label='Küme Merkezleri')
    
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.title("KMeans Clustering - Küme Görselleştirmesi")
    plt.legend(loc='upper left')
    plt.show()
    
Kmeans("Arasınav Notu","Final Sınavı Notu",1,0)
```
